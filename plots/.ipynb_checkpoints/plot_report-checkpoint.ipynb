```python
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = 16, 6
```

```python
n_pods = ['001', '002', '004', '006', '008', '012']
n_versions = ['00', '01']
times = np.zeros((len(n_pods), len(n_versions)))
fmt = '%H:%M:%S'
```

```python
for x, n in enumerate(n_pods):
    for y, v in enumerate(n_versions):
        with open('../logs/async_'+n+'_pods_'+v+'.log', 'r') as file:
            records = []
            still_text = True
            while(still_text):
                line = file.readline()
                if "[LOSS = " in line:
                    if float(line[line.find("[LOSS = ")+8:line.find("][A")]) > 0.2:
                        records.append(line[line.find("[")+1:line.find("]")].split(' ')[3])
                if "I am done!" in line:
                    still_text = False
            tstamp1 = datetime.strptime(records[0], fmt)
            tstamp2 = datetime.strptime(records[-1], fmt)
        
            times[x][y] = (tstamp2-tstamp1).seconds
```

```python
time_one_pod = np.mean(times[0])
```

```python
times2 = np.divide(time_one_pod, np.asarray(times))
```

```python
ax = sns.tsplot(data=times.T, linestyle='--')
ax.lines[-1].set_linestyle('-')
ax.legend()
#plt.yscale('log')
plt.xlabel('number of pods')
plt.ylabel('seconds needed to reach 0.18 as loss')
plt.title('Convergence speed depending on number of pods')
plt.show()
```

```python
ax = sns.tsplot(data=times2.T, linestyle='--')
ax.lines[-1].set_linestyle('-')
ax.legend()
#plt.yscale('log')
plt.xlabel('number of pods')
plt.ylabel('speed-up')
plt.title('Speed-up depending on number of pods compared to one pod')
plt.show()
```
